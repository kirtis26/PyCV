**1) Написать программу на языке Python, реализующую построение трех разных графиков произвольных функций:**
* Файл 1 - три графика в одном окне;
* Файл 2 - три графика в разных окнах.

```python
# подключаем библиотеки
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.linspace(-5, 5) # numpy массив X аргументов х
```

```python
def func1(x):
    """
    Линейная зависимость
    """
    return (0.64 * x - 64)
```

```python
def func2(x):
    """
    Квадратичная зависимость
    """
    return (-(x - 5)**2 - 3 * x - 1)
```

```python
def func3(x):
    """
    Экспоненциальная зависимость
    """
    return np.exp(0.94 * x - 1) - 50
```

```python
fig = plt.figure(dpi=100, figsize=(10, 5)) # создаешь рабочее пространство для отрисовки
ax = fig.add_subplot() # создаешь окно для отрисовки
fig.set(facecolor='darkorange')
ax.set(facecolor='lightgrey')

plt.title('Графики функций в одном окне', color='darkblue', fontsize=14)

plt.plot(X, func1(X), linestyle="-",  linewidth=3, color='blue',  label="Линейная зависимость")
plt.plot(X, func2(X), linestyle="--", linewidth=3, color='red',   label="Квадратичная зависимость")
plt.plot(X, func3(X), linestyle="-.", linewidth=3, color='green', label="Экспоненциальная зависимость")

plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)

plt.grid(color = 'black', linewidth = 1, linestyle = '--')

plt.legend(fontsize=12, loc='upper left', ncol=1, facecolor='oldlace', edgecolor='r', title='Легенда', title_fontsize='14')

plt.show()
```

```python
fig = plt.figure(dpi=100, figsize=(12, 7))

ax_1 = fig.add_subplot(2, 2, 1)
ax_2 = fig.add_subplot(2, 2, 2)
ax_3 = fig.add_subplot(2, 2, 3)

ax_1.plot(X, func1(X), linestyle="-",  linewidth=3, color='blue',  label="Линейная зависимость")
ax_1.grid(color = 'black', linewidth = 1, linestyle = '--')

ax_2.plot(X, func2(X), linestyle="--", linewidth=3, color='red',   label="Квадратичная зависимость")
ax_2.grid(color = 'black', linewidth = 1, linestyle = '--')

ax_3.plot(X, func3(X), linestyle="-.", linewidth=3, color='green', label="Экспоненциальная зависимость")
ax_3.grid(color = 'black', linewidth = 1, linestyle = '--')

ax_1.set(title='Линейная зависимость',         facecolor='lightgrey')
ax_2.set(title='Квадратичная зависимость',     facecolor='lightgrey')
ax_3.set(title='Экспоненциальная зависимость', facecolor='lightgrey')

plt.show()
```

**2) Написать программу на языке Python, реализующую рисование любого объекта, содержащего геометрические фигуры. Область рисования и размер объекта произвольные. При написании программы использовать библиотеку Turtle. Рисунок должен быть отформатирован. Область рисования должна быть подписана (ФИО, группа)**

```python
from tkinter import *
from turtle import *
```

```python
#функция отрисовкa солнца и зайки
def draw():
    
    t.speed(10)
    t.pensize(5)   
    
    t.up()
    #лиса
    #голова
    t.color("#F4A460")
    t.down() # опустить перо вниз
    t.begin_fill() #начать заливку
    t.setheading(15) #задает угол ориентации черепахи
    t.fd(160) #задает дистанцию перемещения
    t.setheading(120) 
    t.fd(100)
    t.setheading(230)
    t.fd(160)
    t.end_fill() #закончить заливку
    
    #нос
    t.setheading(90)
    t.color("#FF0000")
    t.down()
    t.begin_fill() #начать заливку
    t.circle(7)#круг
    t.end_fill() #закончить заливку
    
    #глаза
    t.up()
    t.goto(100,100)
    t.down()
    t.setheading(90)
    t.color("black")
    t.begin_fill() #начать заливку
    t.circle(5)#круг
    t.end_fill() #закончить заливку  
    t.up()
    t.goto(120,70)
    t.down()
    t.setheading(90)
    t.color("black")
    t.begin_fill() #начать заливку
    t.circle(5)#круг
    t.end_fill() #закончить заливку 
    
    #уши
    t.up()
    t.goto(120,120)
    t.color("#F4A460")
    t.down()
    t.begin_fill() #начать заливку
    t.setheading(15)
    t.fd(40)
    t.setheading(230)
    t.fd(40)
    t.setheading(120)
    t.fd(20)
    t.end_fill() #закончить заливку    
    t.up()
    t.goto(140,90)
    t.color("#F4A460")
    t.down()
    t.begin_fill() #начать заливку
    t.setheading(15)
    t.fd(40)
    t.setheading(230)
    t.fd(40)
    t.setheading(120)
    t.fd(20)
    t.end_fill() #закончить заливку 
    
    #тело
    t.up()
    t.goto(140,20)
    t.color("#F4A460")
    t.down()
    t.begin_fill() #начать заливку
    t.setheading(290)
    t.fd(220)
    t.setheading(180)
    t.fd(180)
    t.setheading(65)
    t.fd(230)
    t.end_fill() #закончить заливку 
    
    #хвост
    t.up()
    t.goto(220,-160)
    t.color("black")
    t.down()
    t.begin_fill() #начать заливку
    t.setheading(75)
    t.fd(150)
    t.setheading(150)
    t.fd(30)
    t.setheading(265)
    t.fd(150)
    t.end_fill() #закончить заливку   
    
    #колобок
    #тело
    t.up()
    t.goto(0,70)
    t.down()
    t.setheading(90)
    t.color("#F0E68C")
    t.begin_fill() #начать заливку
    t.circle(50)#круг
    t.end_fill() #закончить заливку
    
    #глаза
    t.up()
    t.goto(-70,90)
    t.down()
    t.setheading(90)
    t.color("black")
    t.begin_fill() #начать заливку
    t.circle(5)#круг
    t.end_fill() #закончить заливку
    
    t.up()
    t.goto(-30,90)
    t.down()
    t.setheading(90)
    t.color("black")
    t.begin_fill() #начать заливку
    t.circle(5)#круг
    t.end_fill() #закончить заливку

    #улыбка
    t.up()
    t.goto(-70,60)
    t.down()
    t.setheading(270)
    t.color("#8B0000")
    t.begin_fill() #начать заливку
    t.circle(20,180)#
    t.end_fill() #закончить заливку
```

```python
t = Turtle() # создает черепаху для отрисовки рисунка
tk = Tk() # создает кнопку для работы с программой
t.screen.setup(660, 660)  # устанавливаем размер рабочего окна
t.screen.bgcolor("lightgrey") # заполняем фон
t.screen.title("Реутова Марта АА-18-05")
tk.configure(bg="#A6E6FA") # конфигурация кнопки (цвет)
btn_draw = Button(tk, text="Рисовать", bg="#C71585", font="Helvetica 10 bold", width=15, command=draw) # создана кнопка
btn_draw.pack() #упаковыват кнопку в рабочем пространтсве
t.screen.mainloop()
```

**3) Написать программу на языке Python, реализующую произвольное движение фотографии. С помощью библиотеки Tkinter отобразить на экране свою фотографию. Не забывайте, что изображение должно быть в формате GIF!**

```python
import random
from tkinter import *
from tkinter import filedialog
```

```python
#функции движения:
def move():
    x=random.randint(-1,1)
    y=random.randint(-1,1)
    c.move(photo, x*10,y*10)
```

```python
tk=Tk()#инициализация области отрисовки
tk.geometry("300x350")#размер области
tk.configure(bg="#FFB6C1")#цвет области
tk.title("Реутова Марта АА-18-05")#название
```

```python
#создание области типа cаnvas для картинки
c=Canvas(tk,width=300,height=300,bg="#FFB6C1")
c.grid(row=0, columnspan=4)#положение области
```

```python
#открытие окна проводника
file_name = filedialog.askopenfilename(filetypes=[('gif images','.gif')])
img = PhotoImage(file=file_name)#получение изображения
img = img.subsample(2,2)#маштабирование

#создание области photo
photo = c.create_image(150,150,image=img) # создает на холсте картинку image размером 150 на 150
#создание кнопки ""
btn_right=Button(tk,text="движение",bg="#C71585",font="Helvetica 12 bold",width=20,command=move)
btn_right.grid(row=1,column=0) #располагаем кнопку во второй половине рабочего пространства для отрисовки
#отрисовка
tk.mainloop()
```
